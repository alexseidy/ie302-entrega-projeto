## Modelos

```
# keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, LeakyReLU
from keras.callbacks import EarlyStopping
from keras import optimizers

class ModeloMLP(Sequential):
    def train(self, xt, yt, xv, yv, 
              hidden_layer_size=100,
              epochs=100, 
              batch_size=64,
              verbose=1,
              patience=5,
              activation='sigmoid'):
        self.add(Dense(hidden_layer_size, 
                       input_dim=xt.shape[1], 
                       activation=activation,
                       kernel_initializer='random_uniform'))
        # self.add(Dense(2, activation='relu'))
        self.add(Dense(1))
        # optimizer = optimizers.RMSprop(lr=0.001, rho=0.9, epsilon=None, decay=0.0)
        self.compile(loss = 'mean_absolute_error',
                    optimizer = 'adam')
        tic = timeit.default_timer()
        self.history = self.fit(xt, yt,
                    epochs = epochs,
                    batch_size = batch_size,
                    verbose = verbose,
                    validation_data = (xv, yv),
                    callbacks = [EarlyStopping(monitor='val_loss', patience=patience)])
        toc = timeit.default_timer()
        print('Tempo de treino (s):', (toc-tic))
    
    def get_history(self):
        return self.history


class ModeloMLP_LeakyReLU(Sequential):
    def train(self, xt, yt, xv, yv, 
              hidden_layer_size=100,
              epochs=100, 
              batch_size=64,
              verbose=1,
              patience=5,
              alpha=0.1):
        self.add(Dense(hidden_layer_size, 
                       input_dim=xt.shape[1], 
                       activation='linear',
                       kernel_initializer='random_uniform'))
        self.add(LeakyReLU(alpha=alpha))
        # self.add(Dense(2, activation='relu'))
        self.add(Dense(1))
        # optimizer = optimizers.RMSprop(lr=0.001, rho=0.9, epsilon=None, decay=0.0)
        self.compile(loss = 'mean_absolute_error',
                    optimizer = 'adam')
        tic = timeit.default_timer()
        self.history = self.fit(xt, yt,
                    epochs = epochs,
                    batch_size = batch_size,
                    verbose = verbose,
                    validation_data = (xv, yv),
                    callbacks = [EarlyStopping(monitor='val_loss', patience=patience)])
        toc = timeit.default_timer()
        print('Tempo de treino (s):', (toc-tic))
    
    def get_history(self):
        return self.history
```

### Erro absoluto médio

```
def calcula_mae(modelo, colunas):
    return np.mean(np.abs(
        modelo.predict(df_valid.loc[:,colunas].values)-df_valid.loc[:,'y'].values))
```

```
print('Erro absoluto médio dos classificadores')
print('MLP(tanh) (stats):', calcula_mae(mlp_tanh_stats, col_stats))
print('MLP(tanh) (mfcc):', calcula_mae(mlp_tanh_mfcc, col_mfcc))
print('MLP(tanh) (full):', calcula_mae(mlp_tanh_full, col_full))
print('MLP(sigm) (stats):', calcula_mae(mlp_sigm_stats, col_stats))
print('MLP(sigm) (mfcc):', calcula_mae(mlp_sigm_mfcc, col_mfcc))
print('MLP(sigm) (full):', calcula_mae(mlp_sigm_full, col_full))
print('MLP(relu) (stats):', calcula_mae(mlp_relu_stats, col_stats))
print('MLP(relu) (mfcc):', calcula_mae(mlp_relu_mfcc, col_mfcc))
print('MLP(relu) (full):', calcula_mae(mlp_relu_full, col_full))
print('LR (stats):', calcula_mae(lr_stats, col_stats))
print('LR (mfcc):', calcula_mae(lr_mfcc, col_mfcc))
print('LR (full):', calcula_mae(lr_full, col_full))
```
