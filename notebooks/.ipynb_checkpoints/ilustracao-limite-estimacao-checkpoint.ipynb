```
## Basic imports
%matplotlib inline
%pylab inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

## Utils
import timeit
import time
from tqdm import tqdm
import pickle
```

```
t_l = np.arange(0,10,.1)
t_s = np.arange(0,6,.1)

with plt.style.context('ggplot'):
    l = 2.0
    figsize(3*l,2*l)
    fig, ax = plt.subplots()
    ax.plot(np.append(0, 10-t_l), '--', label='Tempo até falha (longo)')
    ax.plot(np.append(0, 6-t_s), '--', label='Tempo até falha (curto)')
    plt.legend()
    
    ax2 = ax.twinx()
    plt.grid(False)
    ax2.plot(np.append(1-np.exp(-t_l*.8), 0), label='Tensão de cisalhamento (longo)')
    ax2.plot(np.append(1-np.exp(-t_s*.8), 0), label='Tensão de cisalhamento (curto)')
    # plt.legend()
    plt.ylim([0,2])
    plt.yticks([0,1], labels=['0','$\sigma_{max}$'])

    savefig('ilustracao-limite-estimacao.eps', format='eps', dpi=400)
    savefig('ilustracao-limite-estimacao.png', format='png', dpi=400)
    
```
