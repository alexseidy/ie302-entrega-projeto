## Importando bibliotecas

```
## Basic imports
%matplotlib inline
%pylab inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

## Utils
import timeit
import time
from tqdm import tqdm
import pickle

## Audio features
# import librosa
# import librosa.display
```

```
!cp "$drv/$proj/dados/dataset-preprocessado.pkl" .
with open('dataset-preprocessado.pkl', 'rb') as f:
    dados = pickle.load(f)
    f.close()

x_train = dados['x_train']
y_train = dados['y_train']
y_train_cate = dados['y_train_cate']
x_valid = dados['x_valid']
y_valid = dados['y_valid']
y_valid_cate = dados['y_valid_cate']

del dados
```

```
thresholds = [0.0000, 1.0231, 2.0849, 3.1740, 4.2619,  
              5.3516, 6.4415, 7.5640, 8.8903, 10.8773]
quantize = []
for x in range(len(thresholds)-1):
    quantize.append((thresholds[x+1]+thresholds[x])/2)
quantize.append(thresholds[-1])
```

```
subplot(211)
y = np.array([quantize[p] for p in np.argmax(y_train_cate, axis=1)])
plt.plot(y)
plt.plot(y_train)

subplot(212)
y = np.array([quantize[p] for p in np.argmax(y_valid_cate, axis=1)])
plt.plot(y)
plt.plot(y_valid)
```

## Experimento

```
### Keras (usando o backend TensorFlow)
import tensorflow as tf
from keras import backend as K
from keras.models import Sequential
# from keras.models import Model

# Layers
from keras.layers.normalization import BatchNormalization
from keras.layers.convolutional import Conv1D, MaxPooling1D
from keras.layers.core import Activation, Flatten, Dropout, Dense
from keras.layers import Input

# Optimization
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras import optimizers
from keras.optimizers import Adam

# Regularization
from keras.regularizers import l2
```

```
cnn = Sequential()

cnn.add(Conv1D(filters=1, kernel_size=10, input_shape=(1500,1), 
               activation='tanh',
               kernel_regularizer = l2(l = 0.01)
               ))

cnn.add(MaxPooling1D(pool_size=5))

cnn.add(Conv1D(filters=1, kernel_size=10, 
               activation='tanh',
               kernel_regularizer = l2(l = 0.01)
               ))

cnn.add(MaxPooling1D(pool_size=5))

cnn.add(Flatten())
cnn.add(Dropout(.10)) # resultado ruim

cnn.add(Dense(10, activation='softmax'))

opt = Adam(lr=0.001)
cnn.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

cnn.summary()
```

- (altos tempos de processamento devido ao uso de uma sessão sem aceleração de GPU)
- (restrição de acesso às sessões de GPU do Google Colab)


```
# checkpoint = ModelCheckpoint('cnn_model.keras', 
#                             monitor='val_acc', 
#                             save_best_only=True)

early_stop = EarlyStopping(monitor='val_acc', 
                           patience=10,
                           restore_best_weights=True)

history = cnn.fit(x_train.reshape([-1, 1500, 1]), y_train_cate, 
                  validation_data=(x_valid.reshape([-1, 1500, 1]), y_valid_cate), 
                  epochs=150,
                  callbacks = [early_stop]
                 )
```

```
with plt.style.context('ggplot'):
    figsize(5,3)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.ylabel('Entropia cruzada categórica')
    plt.xlabel('Época')
    plt.legend(['Treino','Validação'])
    savefig('history-cnn-clas.eps', dpi=400, format='eps')
    savefig('history-cnn-clas.png', dpi=400, format='png')

!cp 'history-cnn-clas.'* "$drv/$proj/imagens/"
```

```
with plt.style.context('ggplot'):
    figsize(6,8)
    subplot(211)
    pred = cnn.predict(x_train.reshape(-1,1500,1))
    pred = np.argmax(pred, axis=1)
    pred = np.array([quantize[p] for p in pred])
    plt.plot(pred, label='Valor predito')
    
    y = np.array([quantize[p] for p in np.argmax(y_train_cate, axis=1)])
    plt.plot(y, label='Valor verdadeiro quantizado')

    plt.plot(y_train, label='Valor verdadeiro')
    
    plt.xlabel('Amostra')
    plt.ylabel('Tempo até falha')
    plt.legend()

    print('MAE(train):', np.mean(np.abs(pred.ravel()-y_train)))

    subplot(212)
    pred = cnn.predict(x_valid.reshape(-1,1500,1))
    pred = np.argmax(pred, axis=1)
    pred = np.array([quantize[p] for p in pred])
    plt.plot(pred, label='Valor predito')

    y = np.array([quantize[p] for p in np.argmax(y_valid_cate, axis=1)])
    plt.plot(y, label='Valor verdadeiro quantizado')

    plt.plot(y_valid, label='Valor verdadeiro')
    plt.xlabel('Amostra')
    plt.ylabel('Tempo até falha')
    plt.legend()

    print('MAE(valid):', np.mean(np.abs(pred.ravel()-y_valid)))

    savefig('result-cnn-clas.eps', dpi=400, format='eps')
    savefig('result-cnn-clas.png', dpi=400, format='png')

!cp 'result-cnn-clas.'* "$drv/$proj/imagens/"
```
